# Q3. Perform dimensionality reduction using PCA such that the 95% of the variance is explainedðŸ¡ª2 marks

```python
import numpy as np
from sklearn.decomposition import PCA

# Generate a sample dataset
np.random.seed(10)

# Perform PCA
pca = PCA()
pca.fit(card)

# Determine the number of components needed to explain 95% of the variance
cumulative_variance = np.cumsum(pca.explained_variance_ratio_)
num_components = np.argmax(cumulative_variance >= 0.95) + 1

# Perform PCA with the chosen number of components
pca = PCA(n_components=num_components)
reduced_data = pca.fit_transform(card)

# Print the results
print("Original Data Shape:", card.shape)
print("Reduced Data Shape:", reduced_data.shape)
```

# Q4. Find the optimum value of k for k-means clustering using the elbow method. Plot the elbow curveðŸ¡ª2 marks

```python
from sklearn.cluster import KMeans
 
wsse = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i,  random_state = 10)
    kmeans.fit(card)
    wsse.append(kmeans.inertia_)
plt.plot(range(1, 11), wsse)
# plt.scatter(range(1, 11),wsse)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WSSE')
plt.show()
```

5 is the optimum value of k for k-means clustering using the elbow method

There are two elbow point in 2 and 5 we will consider 5 because after 5 there is no sudden dip in data
